```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import folium
from folium.plugins import HeatMap
```

```python
# 1. Leggi il CSV
df = pd.read_csv(
    r"C:\Users\Dell\OneDrive - Danmarks Tekniske Universitet\Git Hub\StayingAlive\src\restaurants_df.csv"
)  # Cambia col tuo path

# 2. Filtra solo ristoranti attivi e a Copenhagen
df = df[df["active"] == True]
df = df[df["Address"].str.contains("København")]

# 3. Geocodifica con Nominatim
geolocator = Nominatim(user_agent="cph_heatmap")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)


def get_coords(addr):
    try:
        location = geocode(addr)
        if location:
            return pd.Series([location.latitude, location.longitude])
    except:
        return pd.Series([None, None])


df[["lat", "lon"]] = df["Address"].apply(get_coords)

# Rimuovi righe senza coordinate
df = df.dropna(subset=["lat", "lon"])

# 4. Crea la heatmap
mappa = folium.Map(location=[55.6761, 12.5683], zoom_start=12)
heat_data = df[["lat", "lon"]].values.tolist()
HeatMap(heat_data).add_to(mappa)

# Salva la mappa
mappa.save("heatmap_ristoranti.html")
```
